```python
# Import libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None # Adjust the configuration of the plots we will create


# Read in the data

df = pd.read_csv(r'C:\Users\Matt\OneDrive\Desktop\COVID Data Project\movies.csv',encoding='cp1252')
```

```python
# Let's take a look at the data

df.head()
```

```python
# Find missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{}-{}%'.format(col,pct_missing))
```

```python
# Data types for our columns

df.dtypes
```

```python
# Datatype change

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df
```

```python
# Creating new columns to correct year

df['yearcorrect'] = df['released'].astype(str).str[:4]

df
```

```python
df = df.sort_values(by=['gross'],inplace=False, ascending=False)
```

```python
#look at all the data

pd.set_option('display.max_rows',None)
```

```python
#Checking for and dropping duplicates

df['company'].drop_duplicates().sort_values(ascending=False).head()
```

```python
df.head()
```

```python
# Scatter plot with budget vs gross


plt.scatter(x=df['budget'],y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Budget for Film')

plt.ylabel('Gross Earnings')

plt.show()
```

```python
df.head()
```

```python
# Plot Budget vs Gross using Seaborn

sns.regplot(x='budget',y='gross',data=df,scatter_kws={"color":"red"},line_kws={"color":"blue"})
```

```python
# Correlation Workout

df.corr()   #Pearson Correlation
```

```python
df.corr(method='kendall') #Kendall Correlation
```

```python
df.corr(method='spearman') #Spearman Correlation
```

```python
correlation_matrix=df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()

```

```python
# Correlation based on Company

df.head()
```

```python
df_numerized=df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]=df_numerized[col_name].astype('category')
        df_numerized[col_name]=df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
df.head()
```

```python
correlation_matrix=df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
